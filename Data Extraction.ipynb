```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
```

```python
# Testing with 1 page of albums, extracting music data from MusicBoard
url = f"https://api.musicboard.app/v2/albums/popular/?limit=24&offset=216"
ratings_count = []
album_name = []
artist = []
offset = 0
response = requests.get(url)
results = json.loads(response.text)
for entry in results['results']:
    if entry['ratings_count'] > 2500:
        ratings_count.append(entry['ratings_count'])
        album_name.append(entry['title'])
        artist.append(entry['artist']['name'])

df = pd.DataFrame({'Album': album_name,
                   'Artist': artist,
                   'Ratings Count': ratings_count
                  })
df.to_csv('album_list.csv', index=False)

display(df)
```

```python
# Getting album names of albums with 2500+ ratings 
url = f"https://api.musicboard.app/v2/albums/popular/"
ratings_count = []
album_name = []
artist = []
offset = 0
response = requests.get(url)
results = json.loads(response.text)
while (offset < 5000):
    for entry in results['results']:
        if entry['ratings_count'] > 2500:
            ratings_count.append(entry['ratings_count'])
            album_name.append(entry['title'])
            artist.append(entry['artist']['name'])
    offset += 24
    print(offset)
    if results['next']:
        response = requests.get(results['next'])
        results = json.loads(response.text)
    else:
        break

df = pd.DataFrame({'Album': album_name,
                   'Artist': artist,
                   'Ratings Count': ratings_count
                  })
df.to_csv('album_list.csv', index=False)

display(df)
```

```python
new_df = df[['Album','Artist']]
new_df
```

```python
new_df['Album']=new_df['Album'].apply(lambda x:x.lower())
new_df['Artist']=new_df['Artist'].apply(lambda x:x.lower())
new_df
```

```python
#Formatted to go into url slug api point
new_df['Album'] = new_df['Album'].str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
new_df['Album'] = new_df['Album'].str.replace(r"\s+", "-", regex=True)
new_df['Artist'] = new_df['Artist'].str.replace(r"[^a-zA-Z0-9\s]", "", regex=True)
new_df['Artist'] = new_df['Artist'].str.replace(r"\s+", "-", regex=True)
new_df
```

```python
#Getting the album names, average rating, genre, and artist name for all the albums we collected that had 2500+ ratings
ratings_count = []
album_name = []
artist_name = []
average_album_rating = []
artist_rating = []
artist_rating_count = []
genre = []
# test_df = new_df.head(5)
for index, row in new_df.iterrows():
    album = row['Album']  
    artist = row['Artist']
    url = f"https://api.musicboard.app/v2/albums/from_url_slug/?url_slug=/album/{album}/{artist}/"
    try:
        response = requests.get(url)
        results = json.loads(response.text)
        ratings_count.append(results['ratings_count'])
        album_name.append(results['title'])
        artist_name.append(results['artist']['name'])
        average_album_rating.append(results['average_rating'])
        artist_rating.append(results['artist']['average_rating'])
        artist_rating_count.append(results['artist']['ratings_count'])
        list = []
        for item in results['genres']:
            list.append(item['name'])
        str = ', '.join(list)
        genre.append(str)
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for Album: {row['Album']} by Artist: {row['Artist']} - {e}")
    except KeyError as e:
        print(f"Missing data in response for Album: {row['Album']} by Artist: {row['Artist']} - {e}")

ratings_df = pd.DataFrame({'album': album_name,
                        'artist': artist_name,
                        'average album rating': average_album_rating,
                        'album ratings count': ratings_count,
                        'artist rating': artist_rating,
                        'artist rating count': artist_rating_count,
                        'genres': genre,
                    })
ratings_df.to_csv('ratings_data.csv', index = False)
display(ratings_df)

```
