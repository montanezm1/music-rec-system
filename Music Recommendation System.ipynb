```python
df['average album rating'] = pd.to_numeric(df['average album rating'])
df['average album rating'] = df['average album rating'].apply(lambda x: f"{x:.1f}")
df['artist rating'] = pd.to_numeric(df['artist rating'])
df['artist rating'] = df['artist rating'].apply(lambda x: f"{x:.1f}")
df
```

```python
df['artist']=df['artist'].str.replace(',', ' ')
df['artist']=df['artist'].str.replace(' ', '')
df['genres']=df['genres'].str.replace(',', ' ')
df['genres']=df['genres'].str.replace('/', ' ')
df['genres']=df['genres'].str.replace(' ', '')
df
```

```python
# Try adding more tags
df['average album rating'] = df['average album rating'].astype(str)
df['tags']=df['artist']+' '+df['genres']+' '+df['average album rating']
df['tags']
```

```python
df['tags'][0]
```

```python
new_df = df[['album', 'tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.rename(columns={'album': 'title'}, inplace=True)
```

```python
def recommend(music):
    music_index = new_df[new_df['title']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('OK Computer')
```
