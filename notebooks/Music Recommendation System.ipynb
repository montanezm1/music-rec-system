```python
numeric_features = [
    "average album rating",
    "album ratings count",
    "artist rating",
    "artist rating count",
    "num of songs"
]

scaler = MinMaxScaler()
numeric_scaled = scaler.fit_transform(df[numeric_features])


df["release year"] = pd.to_datetime(df["release date"], errors="coerce").dt.year
df["release year"] = df["release year"].fillna(df["release year"].median())
release_scaled = scaler.fit_transform(df[["release year"]])

# Vectorize genres
vectorizer_genres = CountVectorizer(tokenizer=lambda x: x.split(", "), token_pattern=None)
genres_encoded = vectorizer_genres.fit_transform(df["genres"].fillna(""))

# Vectorize label
vectorizer_label = CountVectorizer()
labels_encoded = vectorizer_label.fit_transform(df["record label"].fillna(""))

# Combine everything into feature matrix
import scipy.sparse as sp

feature_matrix = sp.hstack([
    numeric_scaled,
    release_scaled,
    genres_encoded,
    labels_encoded
])
```

```python
import pickle

cosine_sim = cosine_similarity(feature_matrix, feature_matrix)

# save
with open("../data/cosine_sim.pkl", "wb") as f:
    pickle.dump(cosine_sim, f)
```

```python
cosine_df = pd.DataFrame(cosine_sim)
print(cosine_df)
```

```python
import random

def recommend(album_title, df, cosine_sim, top_n=5):
    # Find the index of the given album
    idx = df.index[df['album'].str.lower() == album_title.lower()].tolist()
    if not idx:
        return f"Album '{album_title}' not found."
    idx = idx[0]
    
    # Get similarity scores for this album
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort albums by similarity (highest first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Skip the first one (it's the same album)
    sim_scores = sim_scores[1:]  
    
    # Top N most similar albums
    top_recs = []
    for i, score in sim_scores[:top_n]:
        top_recs.append({
            "album": df.iloc[i]['album'],
            "artist": df.iloc[i]['artist'],
            "cover": df.iloc[i]['cover'],
            "similarity": round(score, 3)  # optional to show similarity score
        })
    
    # Pick 1 random album from all albums
    random_pool = sim_scores[top_n:]
    random_pick = None
    if random_pool:
        i, score = random.choice(random_pool)
        random_pick = {
            "album": df.iloc[i]['album'],
            "artist": df.iloc[i]['artist'],
            "cover": df.iloc[i]['cover'],
            "similarity": round(score, 3)
        }

    return {
        "most_similar": top_recs,
        "random_related": random_pick,
    }


```

```python
# Testing it out
recs = recommend("Un Verano Sin Ti", df, cosine_sim, top_n=5)

print("Most Similar:")
for r in recs["most_similar"]:
    print(f"{r['album']} by {r['artist']} (similarity={r['similarity']})")

print("\nRandom Related:")
print(f"{recs['random_related']['album']} by {recs['random_related']['artist']} (similarity={recs['random_related']['similarity']})")
```
