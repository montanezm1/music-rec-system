```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
```

```python
# Testing with 1 page of albums, extracting music data from MusicBoard
url = f"https://api.musicboard.app/v2/albums/popular/?limit=24&offset=216"
ratings_count = []
album_name = []
artist = []
url_slug = []
offset = 0
response = requests.get(url)
results = json.loads(response.text)
for entry in results['results']:
    if entry['ratings_count'] > 2500:
        ratings_count.append(entry['ratings_count'])
        album_name.append(entry['title'])
        artist.append(entry['artist']['name'])
        url_slug.append(entry['url_slug'])

df = pd.DataFrame({'Album': album_name,
                   'Artist': artist,
                   'Ratings Count': ratings_count,
                   'Url Slug': url_slug,
                  })
df.to_csv('album_list.csv', index=False)

display(df)
```

```python
# Getting album names of albums with 1000+ ratings 
url = f"https://api.musicboard.app/v2/albums/popular/"
ratings_count = []
album_name = []
artist = []
url_slug = []
offset = 0
response = requests.get(url)
results = json.loads(response.text)
while (offset < 5000):
    for entry in results['results']:
        if entry['ratings_count'] >= 1000:
            ratings_count.append(entry['ratings_count'])
            album_name.append(entry['title'])
            artist.append(entry['artist']['name'])
            url_slug.append(entry['url_slug'])
    offset += 24
    print(offset)
    if results['next']:
        response = requests.get(results['next'])
        results = json.loads(response.text)
    else:
        break

df = pd.DataFrame({'Album': album_name,
                   'Artist': artist,
                   'Ratings Count': ratings_count,
                   'Url Slug': url_slug,
                  })
df.to_csv('album_list.csv', index=False)

display(df)
```

```python
#Getting the album names, average rating, genre, and artist name for all the albums we collected that had 2500+ ratings
ratings_count = []
album_name = []
artist_name = []
average_album_rating = []
artist_rating = []
artist_rating_count = []
genre = []
release_date = []
record_label = []
num_songs = []
cover = []
#test_df = new_df.head(5)
for url_slug in new_df:
    url = f"https://api.musicboard.app/v2/albums/from_url_slug/?url_slug={url_slug}"
    try:
        response = requests.get(url)
        results = json.loads(response.text)
        ratings_count.append(results['ratings_count'])
        album_name.append(results['title'])
        artist_name.append(results['artist']['name'])
        average_album_rating.append(results['average_rating'])
        artist_rating.append(results['artist']['average_rating'])
        artist_rating_count.append(results['artist']['ratings_count'])
        release_date.append(results['release_date'])
        record_label.append(results['label'])
        num_songs.append(results['nb_tracks'])
        cover.append(results['cover'])
        list = []
        for item in results['genres']:
            list.append(item['name'])
        str = ', '.join(list)
        genre.append(str)
    except requests.exceptions.RequestException:
        print("Error fetching data for Album:")
    except KeyError:
        print("Missing data in response for Album")

ratings_df = pd.DataFrame({'album': album_name,
                        'artist': artist_name,
                        'average album rating': average_album_rating,
                        'album ratings count': ratings_count,
                        'artist rating': artist_rating,
                        'artist rating count': artist_rating_count,
                        'genres': genre,
                        'release date': release_date,
                        'record label': record_label,
                        'num of songs': num_songs,
                        'cover': cover,
                    })
ratings_df.to_csv('ratings_data.csv', index = False)
display(ratings_df)

```
